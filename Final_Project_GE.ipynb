```python
import os

import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import mapping

import geopandas as gpd
import xarray as xr
import rioxarray as rxr

#import earthpy as et
#import earthpy.spatial as es
#import earthpy.plot as ep

from glob import glob
import os
import urllib

import rasterio as rio
from rasterio.plot import show
import xarray as xr
import rioxarray as rxr


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from rasterio.plot import show_hist
import pandas as pd
```

Band Extraction Function

```python
def band_np(d_fp, pos):
    b = rio.open(d_fp[pos])
    return b.read(1).astype(float)/1000
```

NDVI Calculation Function

```python
def find_ndvi(b_np_1, b_np_2):
    return (b_np_1-b_np_2)/(b_np_1+b_np_2)
```

NDVI Image Generation Function

```python
def fig(a_ndvi):
    return plt.subplots(figsize=(20, 20), sharey=True), plt.imshow(a_ndvi), plt.colorbar()
```

```python
#Example Function:
fig(Donbasj21_ndvi)
```

NDVI Histogram Generation Function

```python
def histo(b_ndvi):
    return show_hist(b_ndvi, bins=50, lw=0.0, stacked=False, alpha=0.3, histtype='stepfilled')
```

```python
#Example Function:
histo(Donbasj21_ndvi)
```
